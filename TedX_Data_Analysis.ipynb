## Average number of views of the whole ted talks

```python
#now we will find the average of all the views in the ted talks happened so far as in dataset to the exact number 
#without decimal, using the trunc function. 
#----------------------------
#trunc():- This function is used to eliminate all decimal parts of the floating-point number
#and return the integer without the decimal part.
#----------------------------


print(math.trunc(sum(views)/len(views)))

#we can also use int typecast directly , its very simple
print(int(sum(views)/len(views)))
```

## View to Like Ratio

```python
#here we will calculate the view to like ratio, maximum, minimum view to like ratio from the dataset
```

```python
#printing the views,likes,view to like ratio

for i in range(len(dt)):
    print(dt[i][3], dt[i][4], int(dt[i][3]/dt[i][4]))
```

```python
# now we have seen the view to like ratio above
#now let us find the max,min,avg ratio
#ill be creating a new list as done in previous method, we shall also see the advanced method later

vtl_ratio = []

for i in range(len(dt)):
    vtl_ratio.append(int(dt[i][3]/dt[i][4]))
```

```python
vtl_ratio
```

```python
#finding the maximum view to like ratio
print("max-ratio", max(vtl_ratio), sep=":")

#finding the minimum view to like ratio
print("min-ratio",min(vtl_ratio),sep=":")

#finding the average view to like ratio
print()
```
